```python
TARGET_CONDITIONS = {
    82272006:  'Common Cold (Acute rhinitis)',
    235595009: 'GERD',
    398057008: 'Tension Headache'
}

conditions_df = nodes_condition[nodes_condition['snomed_id'].isin(TARGET_CONDITIONS.keys())].copy()
conditions_df['label'] = conditions_df['snomed_id'].map(TARGET_CONDITIONS)

print('Selected Conditions from BODHI for MoH file')
conditions_df[['label', 'snomed_id', 'name', 'type_condition', 'triage_level']]
```

```python
meta_cols = ['label', 'snomed_id', 'name', 'concept_type', 'type_condition', 
             'triage_level', 'overall_likelihood']
print('Conditions Metadata')
conditions_df[meta_cols].set_index('label')
```

```python
demo_cols = ['label', 'likelihood_male', 'likelihood_female',
             'likelihood_age_0_1', 'likelihood_age_1_5', 'likelihood_age_6_12',
             'likelihood_age_13_18', 'likelihood_age_19_30', 'likelihood_age_30_45',
             'likelihood_age_45_60', 'likelihood_age_60_plus']

demo_df = conditions_df[demo_cols].set_index('label')
demo_df.columns = [c.replace('likelihood_', '') for c in demo_df.columns]

print('Demographic Likelihood Scores')
demo_df.T
```

```python
# Likelihood text to numeric mapping for ranking
LIKELIHOOD_MAP = {
    'very_high': 5,
    'high': 4,
    'medium': 3,
    'low': 2,
    'rare': 1,
    'zero': 0
}

def build_symptom_table(condition_snomed_id, condition_label):
    """Build a full symptom table for a given condition."""
    edges = edges_present_in[edges_present_in['condition_snomed_id'] == condition_snomed_id].copy()
    merged = edges.merge(
        nodes_symptom[['uuid', 'root_snomed_name', 'name', 'triage_level']],
        left_on='symptom_uuid', right_on='uuid',
        how='left'
    )


    merged['score_symptom_given_condition'] = merged['likelihood_symptom_given_condition'].map(LIKELIHOOD_MAP)
    merged['score_condition_given_symptom'] = merged['likelihood_condition_given_symptom'].map(LIKELIHOOD_MAP)
    merged['strong_predictor'] = merged['strong_predictor'].fillna(0).astype(int)
    merged = merged.sort_values('score_condition_given_symptom', ascending=False)
    
    display_cols = ['root_snomed_name', 'name', 
                    'likelihood_symptom_given_condition', 'score_symptom_given_condition',
                    'likelihood_condition_given_symptom', 'score_condition_given_symptom',
                    'strong_predictor', 'triage_level']
    
    result = merged[display_cols].copy()
    result.columns = ['Root Symptom', 'Symptom Variant', 
                      'P(S|C) Label', 'P(S|C) Score',
                      'P(C|S) Label', 'P(C|S) Score',
                      'Strong Predictor', 'Symptom Triage']
    
    return result.reset_index(drop=True)


symptom_tables = {}
for snomed_id, label in TARGET_CONDITIONS.items():
    symptom_tables[label] = build_symptom_table(snomed_id, label)
    print(f'\n{"=" * 60}')
    print(f'  {label}  —  {len(symptom_tables[label])} symptoms linked')
    strong = symptom_tables[label]['Strong Predictor'].sum()
    print(f'  Strong predictors: {strong}')
    print(f'{"=" * 60}')
    display(symptom_tables[label])
```

```python
treated_by = edges_treated_by[
    edges_treated_by['condition_snomed_id'].isin(TARGET_CONDITIONS.keys())
].merge(
    nodes_speciality, left_on='speciality_id', right_on='id', how='left'
)
treated_by['condition_label'] = treated_by['condition_snomed_id'].map(TARGET_CONDITIONS)
treated_by['weight_score'] = treated_by['weight'].map(LIKELIHOOD_MAP)

speciality_pivot = treated_by.pivot_table(
    index='name', columns='condition_label', 
    values='weight', aggfunc='first'
).fillna('—')

print('Speciality Referral Matrix')
print('(weight = how relevant this speciality is for the condition)\n')
speciality_pivot
```

```python
treated_by[['condition_label', 'name', 'weight', 'weight_score']].sort_values(
    ['condition_label', 'weight_score'], ascending=[True, False]
).rename(columns={
    'condition_label': 'Condition',
    'name': 'Speciality',
    'weight': 'Relevance',
    'weight_score': 'Score'
}).reset_index(drop=True)
```

```python
summary_rows = []
for snomed_id, label in TARGET_CONDITIONS.items():
    table = symptom_tables[label]
    cond = conditions_df[conditions_df['snomed_id'] == snomed_id].iloc[0]
    specs = treated_by[treated_by['condition_snomed_id'] == snomed_id]
    primary_spec = specs.sort_values('weight_score', ascending=False).iloc[0]['name'] if len(specs) > 0 else 'N/A'
    
    summary_rows.append({
        'Condition': label,
        'SNOMED ID': snomed_id,
        'BODHI Name': cond['name'],
        'Type': cond['type_condition'],
        'Triage': cond['triage_level'],
        'Overall Likelihood': cond['overall_likelihood'],
        'Total Symptoms': len(table),
        'Strong Predictors': int(table['Strong Predictor'].sum()),
        'Unique Root Symptoms': table['Root Symptom'].nunique(),
        'Linked Specialities': len(specs),
        'Primary Speciality': primary_spec
    })

summary_df = pd.DataFrame(summary_rows).set_index('Condition')
print('Final Summary')
summary_df.T
```
